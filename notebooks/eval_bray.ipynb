```python
data = pl.read_csv('../_data/gigadb/gigadb.csv')
data
```

```python
metadata_keywords = {
    "Location", "Number", "Object", "Count", "EulerNumber", 
    "FileName", "PathName", "URL", "Parent", "Children"
}

feature_columns = [
    col for col in data.columns
    if col.startswith(("Image_", "Cells_", "Cytoplasm_", "Nuclei_"))
    and not any(kw in col for kw in metadata_keywords)
]

# Dynamically grab all metadata columns, explicitly leaving out 'Metadata_Site'
metadata_columns = [
    col for col in data.columns 
    if col not in feature_columns and col != "Metadata_Site"
]

# 2. STEP 1: Aggregate Site-Level to Well-Level
# We use the median across sites to suppress single-site imaging anomalies
well_df = data.group_by(metadata_columns).agg([
    pl.col(col).median().alias(col) for col in feature_columns
])

# 3. STEP 2: Perform Robust MAD Normalization on the Well-Level data
normalization_expressions = []

for col in feature_columns:
    # Isolate DMSO wells on the fly for calculation
    dmso_values = pl.when(pl.col("Metadata_cpd_name") == "DMSO").then(pl.col(col)).otherwise(None)
    
    # Calculate DMSO median per plate
    dmso_median = dmso_values.median().over("Metadata_Plate")
    
    # Calculate DMSO Median Absolute Deviation (MAD) per plate
    absolute_deviation = (dmso_values - dmso_median).abs()
    dmso_mad = absolute_deviation.median().over("Metadata_Plate")
    
    # Constant to make MAD consistent with standard deviation 
    scaled_mad = dmso_mad * 1.4826
    
    # Build the robust normalization expression
    robust_expr = (
        pl.when(scaled_mad == 0)
        .then(0.0)
        .otherwise((pl.col(col) - dmso_median) / scaled_mad)
        .alias(col)
    )
    normalization_expressions.append(robust_expr)

# Apply normalization across the well-level dataframe
normalized_well_df = well_df.with_columns(normalization_expressions)

# 4. STEP 3: Rename feature columns to anonymized index format
rename_mapping = {col: f"feat_{idx}" for idx, col in enumerate(feature_columns)}
final_well_df = normalized_well_df.rename(rename_mapping)
```
