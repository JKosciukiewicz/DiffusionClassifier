```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
```

```python
transform = transforms.Compose([
    transforms.ToTensor()
])

# 2. Apply the transform to the dataset
train_dataset_cifar = datasets.CIFAR10(
    root='../_data/', 
    train=True, 
    download=True, 
    transform=transform
)
```

```python
train_loader = DataLoader(train_dataset_cifar, batch_size=4, shuffle=True)
```

```python
images, labels = next(iter(train_loader))
```
