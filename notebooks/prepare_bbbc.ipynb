```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("../data/BBBC021/BBBC021_dataset_complete.csv")
data
```

```python
labels = pd.read_csv("../data/BBBC021/BBBC021_labels.csv")
labels
```

```python
train_cal_val_df, test_df = train_test_split(data, test_size=0.2, random_state=42)
```

```python
train_cal_df, val_df = train_test_split(
    train_cal_val_df, test_size=0.25, random_state=42
)
```

```python
train_df, cal_df = train_test_split(train_cal_df, test_size=0.20, random_state=42)
```

```python
train_df["split"] = "train"
val_df["split"] = "val"
test_df["split"] = "test"
cal_df["split"] = "cal"
```

```python
data_with_split = pd.concat([train_df, val_df, test_df, cal_df]).reset_index(drop=True)
```

```python
data_with_split = data_with_split.sort_values(
    by="Unnamed: 0", ascending=True, ignore_index=True
)
```

```python
data_with_split
```

```python
# data_with_split.to_csv('../data/BBBC021/BBBC021_dataset_complete_one_fold.csv', index=False)
```

```python
feature_columns = [
    col
    for col in data_with_split.columns
    if col.startswith("feat_") or col == "Unnamed: 0" or col == "split"
]
feature_columns
```

```python
moa_columns = [col for col in labels.columns if col not in ["meta_path"]]
moa_columns
```

```python
labels = labels[moa_columns]
labels
```

```python
data_with_split = data_with_split[feature_columns]
```

```python
merged_df = pd.merge(data_with_split, labels)
```

```python
merged_df
```

```python
merged_df = merged_df.drop(["Unnamed: 0"], axis=1)
merged_df
```
