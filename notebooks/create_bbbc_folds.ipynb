```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold
```

```python
df = pd.read_csv("../data/BBBC021/BBBC021_labels.csv")
df
```

```python
df["fold"] = -1
df
```

```python
# 2. Gather multi-label columns
moa_cols = [c for c in df.columns if c.startswith("moa_")]
y = df[moa_cols].astype(int)

# 3. Create a new string label representing combinations
df["combo"] = y.apply(lambda row: "_".join(row.astype(str)), axis=1)

# 4. Add fold column
df["fold"] = -1
skf = StratifiedKFold(n_splits=4, shuffle=True, random_state=42)

# 5. Assign folds based on combo
for fold, (_, val_idx) in enumerate(skf.split(df, df["combo"])):
    df.loc[val_idx, "fold"] = fold

# 6. Cleanup if needed
df = df.drop(columns=["combo"])
```
